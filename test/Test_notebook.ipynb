```python
import pandas as pd  
```

# **Creating datasets**

```python
# assign data of lists.  
data = {'index': ['USDA', 'USDA', 'USDA', 'USDA'], 'Product': ["tomato raw", "Tomato, ripe, raw, origin unknown", "Tomato, dried", "Tomato, Danish, ripe, raw"],
       'Measuring unit': ["g", "g", "g", "g"], "Amount of": [100, 100, 100, 100], "VITB6_mg":[0.02,0.6, 0.45, 0.01]}  
  
# Create DataFrame  
df_usda = pd.DataFrame(data)  
  
# Print the output.  
df_usda
```

```python
# assign data of lists.  
data = {'index': ['Frida', 'Frida', 'Frida', 'Frida'], 'Product': ["tomato raw", "Tomato, ripe, raw, origin unknown", "Tomato, dried", "Tomato, Danish, ripe, raw"],
       'Measuring unit': ["g", "g", "g", "g"], "Amount of": [100, 100, 100, 100], "VITB6_mg":[0.02,1.6, 0.45, 0.01]}  
  
# Create DataFrame  
df_frida = pd.DataFrame(data)  
  
# Print the output.  
df_frida
```

```python
# assign data of lists.  
data = {'index': ['NEVO', 'NEVO', 'NEVO', 'NEVO'], 'Product': ["tomato raw", "Tomato, ripe, raw, origin unknown", "Tomato, dried", "Tomato, Danish, ripe, raw"],
       'Measuring unit': ["g", "g", "g", "g"], "Amount of": [100, 100, 100, 100], "VITB6_mg":[0.03,0.2, 0.37, 0.01]}  
  
# Create DataFrame  
df_NEVO = pd.DataFrame(data)  
  
# Print the output.  
df_NEVO 
```

```python
#Concatenation the subsets
frames = [df_NEVO, df_frida, df_usda]
df_exemple = pd.concat(frames)
```

# Examples of questions we may have - (data fake)

**Table of all datasets together (NEVO, Frida, USDA). In this example, we have just three different products and just one nutrient, VIT B6**

```python
df_exemple
```

**Sorting the amount of VIT B6: We can check which "tomato" have the biggest and lowest amount**

```python
df_exemple.sort_values("VITB6_mg", ascending=False)
```

**I'd like to compare just tomato raw across the different databases:**

```python
df_exemple.loc[df_exemple["Product"] == "tomato raw"].sort_values("VITB6_mg", ascending=False)
```

**Considering "Tomato raw" what is the mean of VITB6 among the data?**

```python
tomato_raw = df_exemple.loc[df_exemple["Product"] == "tomato raw"]
tomato_raw
```

```python
mean = tomato_raw[["VITB6_mg"]].mean()
mean
```

**The recommended daily amount of vitamin B-6 for adults 50 and younger is 1.3 milligrams. Which tomatoes have more than the recommended?** 

```python
df_exemple.loc[df_exemple["VITB6_mg"] >= 1.3]
```
